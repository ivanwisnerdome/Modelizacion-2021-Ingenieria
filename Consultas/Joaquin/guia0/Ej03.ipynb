## Problema 3

a) Un paracaidista de 70 kg. salta desde un globo aerostático fijo. El coeficiente de resistencia del aire es de
aproximadamente 12.5 kg/seg. (tome la aceleración de la gravedad como 10 m/seg.2). Utilizando la expresión
exacta evalúe la velocidad del paracaidista cada 2 seg. después del salto y calcule su velocidad límite. Grafique
el resultado.
b) Repita el gráfico con la solución discreta, tomando intervalos de 2 seg., 1 seg. y 0.5 seg.
c) Haga el mismo cálculo pero suponiendo ahora que la fuerza de rozamiento se debe expresar como F = – c' v2
(tome c' = 0.25 kg/m).

```python
import numpy as np
import matplotlib.pyplot as plt

m = 70      #Masa del paracaidista [kg]
r = 12.5    #Coeficiente de resistencia del aire [kg/s]
c = 0.25    #[kg/m]
#r = -c*(v**2) 
g = 10      #Aceleracion de la gravedad [m/s^2]

delta_t = 2 #Intervalo de discretizacion
t = np.arange(0, 40, delta_t)
```

#### Ecuacion diferencial (continua): caida del paracaidista:
$m\cdot \frac{dv}{dt} = mg - γv$
  --> cuya solucion es: $ v = \frac{gm}{γ} (1 - e^-(γ/m)t)$ y existe una velocidad limite $ v = \frac{gm}{γ}$
  
#### Ecuacion diferencial (discretizada): caida del paracaidista:
$\frac{v_{i+1} -  v_{i}}{Δt} = g - \frac{γ}{m}v_i \Rightarrow   v_{i+1} = v_i + (g + \frac{γ}{m}v_i)Δt$

```python
v = ((g*m)/r)*(1 - np.exp(-(r/m)*t))   #continuo
v_lim = ((g*m)/r)

vii = np.zeros(len(t),)
for i in range(1, len(t)):
    vi = ((g*m)/r)*(1 - np.exp(-(r/m)*t[i]))
    vii[i] = vi + (g - (r/m)*vi)*(t[i]-t[i-1])


plt.plot(t, v, label='Solucion continuna')
plt.plot(t, vii, label='Solucion discreta')
plt.legend()
print(f'La velocidad limite es: {v_lim} m/s')
plt.xlim(0)
plt.ylim(0)
```

### RESULTADOS:

Δt=2 s 
 
Δt=1 s
 
Δt=0.5 s 

Tomando intervalos Δt mas pequeños la solucion discreta se aproxima mejor a la exacta.
