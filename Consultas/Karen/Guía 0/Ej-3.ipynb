aproximadamente 12.5 kg/seg. (tome la aceleración de la gravedad como 10 m/seg.2). Utilizando la expresión
exacta evalúe la velocidad del paracaidista cada 2 seg. después del salto y calcule su velocidad límite. Grafique
el resultado.
b) Repita el gráfico con la solución discreta, tomando intervalos de 2 seg., 1 seg. y 0.5 seg.
c) Haga el mismo cálculo pero suponiendo ahora que la fuerza de rozamiento se debe expresar como F = – c' v2

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,40,2)
vlim=10*70/12.5
print(vlim)
```

```python
plt.rc('figure', figsize=(20,20))
plt.rc('axes', labelsize=24)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
plt.rc('legend', fontsize=22)
```

```python
v=[0]
v_new = [0]
for i in range(20-1):
    #MDF_COMMENT ojo con el 0-1 = -1 !
    j=v[i-1]+(10-((12.5*v[i-1])/70))*2
    v.append(j)
for i in range(1, 20):
    j_new=v_new[i-1]+(10-((12.5*v_new[i-1])/70))*2
    v_new.append(j_new)
plt.plot(t,v, label='condicion inicial mal tomada')
plt.plot(t,v_new, label='con rango desde 1')
#MDF_COMMENT cuestiones de forma
plt.ylabel('$V_Y$')
plt.xlabel('t')
plt.legend()
```

b) Repita el gráfico con la solución discreta, tomando intervalos de 2 seg., 1 seg. y 0.5 seg. 

```python
i=0
p=2
for j in range(3):
    t=np.arange(0,40,p)
    v=[0]
    for i in range(int(40/p)-1):
        v.append(v[i]+(10-(12.5*v[i]/70))*p)
    #MDF_COMMENT : indicar que gráfico es cada uno !
    plt.plot(t,v, label = 'dt = {:.2f}'.format(p))
    p/=2
plt.legend()
```
