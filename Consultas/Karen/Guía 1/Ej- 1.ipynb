# Problema 1

Se tiene una placa cuadrada de aluminio cuyos bordes están en las condiciones que se indican en la
figura: tres lados a temperaturas fijas y otro, el inferior, aislado térmicamente (flujo de calor igual a
cero). Encuentre, utilizando el método de diferencias finitas, los valores de temperatura y flujo de
calor para cada punto. Utilice primero la grilla señalada y luego generalice el problema a un número
variable de nodos en cada eje de coordenadas. Haga un estudio del tiempo de cálculo en función del
tamaño del sistema.


Para generalizarlo más, falta cambiar el algunos x por y (los diferenciales)

```python
import numpy as np
import matplotlib.pyplot as plt
import copy as cp
```

```python
# Condiciones de contorno
Tu=100
Td=0
Tl=75
Tr=50
```

```python
def temp(Tl, Tu, Tr, Td, nx,ny, q=False, prints=True):
    '''x= l,u, r, d = left, up, right, down
    Si hay algún flujo de calor (o no lo hay) en algún extremo, poner 
    Tx=Q  (Q:calor) y agregar q="x" o q= ["x1", "x2"]'''
    
    b=np.zeros(nx*ny)
    A=np.identity(nx*ny)
    B=ny/nx
    x=1/(nx+1)
    y=1/(ny+1)

    b[0]=(Td+Tl)/2
    b[(nx-1)]=(Td+Tr)/2
    b[(ny-1)*nx]=(Tu+Tl)/2
    b[nx-1 + (ny-1)*nx]=(Tu+Tr)/2

    for i in range(1,nx-1):
        b[i]=Td
        b[i + (ny-1)*nx]=Tu
    for j in range(1,ny-1):
        b[0 + j*nx]=Tl
        b[nx-1 + j*nx]=Tr


    for i in range(1,nx-1):
        for j in range(1,ny-1):
            k= i + j*nx
            A[k,k-1]=1
            A[k,k-nx]=B**2
            A[k,k]=-2*(1+B**2)
            A[k,k+1]=1
            A[k,k+nx]=B**2
    
    # Si hay flujo de calor en algún extremo, se modifican:
    # los valores de la matriz A y de b en ese extremo

    if q!=False:
        
        if "u" in q:
                if prints==True:
                    print("q up")
                for i in range(1,nx-1):
                    b[i + (ny-1)*nx]=Tu*2*x

                for i in range(1,nx-1):

                        k= i + (ny-1)*nx
                        A[k,k-1]=1
                        A[k,k-nx]=2*B**2 -2*x*Tu
                        A[k,k]=-2*(1+B**2)
                        A[k,k+1]=1
                b[(ny-1)*nx]=Tu
                b[nx-1 + (ny-1)*nx]=Tu

        if "d" in q:
                if prints==True:
                    print("q down")
                for i in range(1,nx-1):
                    b[i]=Td*2*x
                for i in range(1,nx-1):
                        k= i
                        A[k,k-1]=1
                        A[k,k+nx]=2*(B**2) -2*x*Td
                        A[k,k]=-2*(1+B**2)
                        A[k,k+1]=1
                b[0]=Tl
                b[(nx-1)]=Tr


        if "l" in q:
                if prints==True:
                    print("q left")
                for j in range(1,ny-1):
                    b[0 + j*nx]=Tl*2*x

                for j in range(1,ny-1):
                    k= j*nx
                    A[k,k+1]=2 -2*x*Tl
                    A[k,k-nx]=B**2
                    A[k,k]=-2*(1+B**2)
                    A[k,k+nx]=B**2
                b[0]=Tl
                b[(ny-1)*nx]=Tl

        if "r" in q:
                if prints==True:
                    print("q right")

                for j in range(1,ny-1):
                    b[nx-1 + j*nx]=Tr*2*x

                for j in range(1,ny-1):
                        k=nx-1 + j*nx
                        A[k,k-1]=2 -2*x*Tr
                        A[k,k-nx]=B**2
                        A[k,k]=-2*(1+B**2)
                        A[k,k+nx]=B**2
                b[(nx-1)]=Tr
                b[nx-1 + (ny-1)*nx]=Tr
                
    t= np.linalg.solve(A,b)
    
    # Matriz temperaturas halladas
    T=np.zeros([nx,ny])
    for i in range(nx):
        for j in range(ny):
            T[j,i]=t[i+nx*j]

    # Flujos
    Fx = np.zeros([nx,ny]) 
    Fy = np.zeros([nx,ny])

    for i in range(nx-1):
        for j in range(ny-1):
            Fx[j,0]=(T[j,1]-T[j,0])/x
            Fx[j,i]=(T[j,i+1]-T[j,i-1])/(2*x)
            Fx[j,nx-1]=(T[j,nx-1]-T[j,nx-2])/x

    for i in range(nx-1):
        for j in range(ny-1):
            Fy[0,i]=(T[1,i]-T[0,i])/x
            Fy[j,i]=(T[j+1,i]-T[j-1,i])/(2*x)
            Fy[ny-1,i]=(T[ny-1,i]-T[nx-2,i])/x

    return A, b, t, T, Fx, Fy
```

```python
n=10
nx=n
ny=n
A, b, t, T, Fx, Fy= temp(Tl, Tu, Tr, Td, nx, ny, q="d")
            
x = np.linspace(0,nx-1,nx).astype(int) #astype para definir numeros enteros
y = np.linspace(0,ny-1,ny).astype(int)
X,Y = np.meshgrid(x,y)

plt.rc('figure', figsize = (10,5))
plt.contourf(T)
plt.colorbar()        
plt.streamplot(X,Y,-Fx,-Fy,color='k')
```

# Análisis de tiempo de corridas

```python
import time 
import numpy as np
import matplotlib.pyplot as plt

Tu=100
Td=0
Tl=75
Tr=50

tiempo=[]
n=np.arange(1,100)

for i in n:
    ti=time.time()
    temp(Tl, Tu, Tr, Td, i,i, q="d", prints=False)
    tf=time.time()
    tiempo.append(tf-ti)
    
plt.plot(n, tiempo)
    
```
